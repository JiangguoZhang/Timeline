# Timeline figure generation

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Timeline data
events = [
    ("Collaboration: Finalize figures, draft manuscript", "2024-03-01", "2024-04-30"),
    ("Thesis: Storyboard, Turing test, methods section", "2024-03-01", "2024-03-31"),
    ("Thesis: Revise manuscript, draft literature review", "2024-04-01", "2024-04-30"),
    ("Thesis: Submit paper", "2024-05-01", "2024-05-31"),
    ("Thesis: Finalize thesis draft", "2024-05-01", "2024-05-31"),
    ("Quantlab Internship", "2024-06-04", "2024-08-16"),
    ("Thesis: Draft thesis defense", "2024-08-17", "2024-09-02"),
    ("Tik Tok Internship", "2024-09-03", "2024-11-22"),
    ("Thesis: Finalize thesis presentation", "2024-09-03", "2024-11-22"),
    ("Apply for jobs", "2024-09-03", "2024-11-22"),
    ("Thesis defense", "2024-11-23", "2024-12-09"),
    ("Graduation", "2024-12-10", "2024-12-10")
]

# Convert dates to datetime objects
for i, (task, start, end) in enumerate(events):
    start_date = datetime.strptime(start, "%Y-%m-%d")
    end_date = datetime.strptime(end, "%Y-%m-%d")
    events[i] = (task, start_date, end_date)

# Plotting
fig, ax = plt.subplots(figsize=(10, 3))

# Timeline
start_date = datetime(2024, 3, 1)
end_date = datetime(2024, 12, 31)
ax.set_xlim(start_date, end_date)

# Add events to timeline
for i, (task, start, end) in enumerate(events):
    ax.plot([start, end], [i] * 2, marker = '|', markersize = 10, linewidth=5)
    ax.text(start, i - 0.2, task, ha='right', va='center', fontsize=9)

# Formatting
ax.set_yticks([])
ax.set_title('Graduation Plan Timeline')
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
plt.xticks(rotation=45, ha='right')
ax.grid(True, which='major', linestyle='--', color='grey')

plt.tight_layout()
plt.show()
```
